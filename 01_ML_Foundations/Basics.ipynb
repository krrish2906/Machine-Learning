## **MatplotLib**

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x,y,z,sep="\n\n")
```

```python
# Plotting Sine wave
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("Sine Wave")
plt.show()
```

```python
# Plotting Cosine wave
plt.plot(x,z)
plt.xlabel("x")
plt.ylabel("cos(x)")
plt.title("Cosine Wave")
plt.show()
```

```python
# plotting a parabola
x = np.linspace(-10,10,100)
y = x**2
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y = x^2")
plt.title("Parabola")
plt.show()
```

```python
# multiple curves in same plot
x = np.linspace(0,10,50)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

```python
# Bar chart
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English', 'Hindi', 'Marathi', 'Gujrathi', 'Haryanvi']
people = np.random.randint(10,100,5)
ax.bar(languages, people)
plt.xlabel("Languages")
plt.ylabel("No. of People")
plt.title("Popularity of Languages")
plt.show()
```

```python
# Pie chart
fig = plt.figure()
ax = fig.add_axes([0,0,0.5,0.5])
languages = ['English', 'Hindi', 'Marathi', 'Gujrathi', 'Haryanvi']
people = np.random.randint(10,100,5)
ax.pie(people, labels=languages, autopct='%1.1f%%')
plt.show()
```

```python
# Scatter plot
x = np.linspace(0,10,50)
y = np.sin(x)
z = np.cos(x)

fig = plt.figure()
ax = fig.add_axes([0,0,0.5,0.5])
ax.scatter(x, y, color='red', marker='.')
ax.scatter(x, z, color='blue', marker='.')
plt.show()
```

```python
# 3D plot
fig = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(250)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x, y, z, c=z, cmap='Reds')
plt.show()
```
