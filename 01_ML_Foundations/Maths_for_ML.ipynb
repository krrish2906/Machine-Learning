Projection of vector A on Vector B

```python
def projection(a, b):
    magnitude_of_b = np.sqrt(sum(b**2))
    numerator = np.dot(a,b)
    proj = (numerator / (magnitude_of_b**2)) * b
    return proj
```

```python
a = np.array([5,4])
b = np.array([8,-6])

print("Projection of a on b: ", projection(a,b))
print("Projection of b on a: ", projection(b,a))
```

```python
v1 = np.array([5,6,-3])
v2 = np.array([-1,2,4])

print("Projection of v1 on v2: ", projection(v1,v2))
print("Projection of v2 on v1: ", projection(v2,v1))
```

Matrix Multiplications


> Two matrices can be multiplied only if the number of columns in the first matrix is equal to the number of rows in the second matrix.


> The resultant matrix will have the same number of rows as the first matrix and the same number of columns as the second matrix.




```python
mat1 = np.array([[4,3,9],[4,3,0],[6,7,5]])
mat2 = np.array([[5,0,3],[0,5,8],[1,4,3]])
result = np.matmul(mat1,mat2)
print(result)
```

```python
a = np.array([[2,3,7],[3,1,4]])
b = np.array([[7,4],[-1,3],[-3,1]])
result = np.matmul(a,b)
print(result)
```

1.   ***Normal Distribution:*** A normal distribution is an arrangement of a data set in which most of the data points lie in the middle of the range and the rest taper off symmetrically toward either extreme.

***Poisson Distribution:***


1.   Poisson Distribution is a probability distribution that measures how many times an event is likely to occur within a specified period of time.
2.   Poisson distribution is used to understand independent events that occur at a constant rate within a given interval of time.

Examples of Poisson Distribution:

1.   Number of accidents occurring in a city from 6 pm to 10 pm.
2.   Number of Patients arriving in an Emergency Room between 10 pm to 12 pm.
3.   How many views does your blog gets in a day.
