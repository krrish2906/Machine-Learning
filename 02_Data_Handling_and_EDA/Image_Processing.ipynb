## **2.) Resizing an image file using Pillow lib**

```python
from PIL import Image
```

```python
img = Image.open("/content/cat.jpg")
img_resized = img.resize((200, 200))
```

```python
img_resized.save("/content/cat_resize.jpg")
```

```python
img2 = mpimg.imread("/content/cat_resize.jpg")
img_plot = plt.imshow(img2)
plt.show()
```

## **3.) Converting RGB image file to Grayscale image using OpenCV**

```python
import cv2
```

```python
img = cv2.imread("/content/cat.jpg")
type(img)
```

```python
img.shape
```

```python
grayscale = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
type(grayscale)
```

```python
grayscale.shape
```

```python
# cv2.imshow() displays the image (but this doesn't works in Google Colab)
from google.colab.patches import cv2_imshow
cv2_imshow(grayscale)
```

```python
cv2.imwrite("/content/cat_grayscale.jpg", grayscale)
```
