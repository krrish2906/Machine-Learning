# **Breast Cancer**

**Importing Dependencies**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

**Loading Dataset and Data Description**

```python
ds = load_breast_cancer()
```

```python
df = pd.DataFrame(data=ds.data, columns=ds.feature_names)
df.shape
```

```python
df['label'] = ds.target
df.shape
```

**Data Analysis**

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['label'].value_counts()
```

```python
# 1 ---> Benign, 0 ---> Malignant
df.groupby('label').mean()
```

**Separating input features (x) and output label (Y)**

```python
X = df.drop(columns='label', axis=1)
Y = df['label']
```

**Splitting Dataset into Training Testing Dataset**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=23)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**Data Standardization**

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

**Building Neural Network**

```python
import tensorflow as tf
from tensorflow import keras
tf.random.set_seed(3)
```

**Setting up the layers of Neural Network**

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

**Training the Neural Network**

```python
hist = model.fit(X_train, Y_train, validation_split=0.1, epochs=10)
```

**Visualization of Accuracy and Loss**

```python
plt.plot(hist.history['accuracy'])
plt.plot(hist.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.legend(['train', 'val'], loc='lower right')
plt.show()
```

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epochs')
plt.legend(['train', 'val'], loc='upper right')
plt.show()
```

**Model Accuracy Evaluation on Testing Dataset**

```python
loss, accuracy = model.evaluate(X_test, Y_test)
print(f'Test Accuracy: {accuracy:.2%}')
```

```python
Y_test_pred = model.predict(X_test)
print(Y_test_pred[0])
```

```python
Y_pred_labels = [np.argmax(i) for i in Y_test_pred]
print(Y_pred_labels[0])
```

```python
print(np.array(Y_pred_labels))
```

**Making a Predictive System**

```python
input_data = (12.77,21.41,82.02,507.4,0.08749,0.06601,0.03112,0.02864,0.1694,0.06287,0.7311,1.748,5.118,53.65,0.004571,0.0179,0.02176,0.01757,0.03373,0.005875,13.75,23.5,89.04,579.5,0.09388,0.08978,0.05186,0.04773,0.2179,0.06871)
input_data = np.asarray(input_data)
input_data = input_data.reshape(1,-1)
input_data = scaler.transform(input_data)

prediction = model.predict(input_data)
print(prediction)

prediction_labels = [np.argmax(prediction)]
print(np.asarray(prediction_labels))

if prediction_labels[0] == 0:
    print('The breast cancer is Malignant')
else:
    print('The breast cancer is Benign')
```
