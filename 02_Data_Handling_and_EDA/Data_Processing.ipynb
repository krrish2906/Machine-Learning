## **Handling Missing values**


Placement Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/Datasets/Placement_Dataset.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
fig, ax = plt.subplots(figsize=(4,4))
sns.distplot(df.salary)
```

1. Imputation Method

```python
# Filling missing values with mean
df['salary'].fillna(df['salary'].mean(), inplace=True)
```

```python
# Filling missing values with median
df['salary'].fillna(df['salary'].median(), inplace=True)
```

```python
# Filling missing values with mode
df['salary'].fillna(df['salary'].mode(), inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

## **Handling Uneven Class Distribution Dataset**

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Datasets/credit_data.csv")
df.shape
```

```python
df.head()
```

```python
df.tail()
```

```python
# Value counts of column 'Class'
df.Class.value_counts()
```

There's a very large imbalance in both the classes

```python
legit = df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Under Sampling

```python
legit_new_sample = legit.sample(n=492)
print(legit_new_sample.shape)
```

```python
# Concatenate legit_new_sample with fraud samples to get balanced dataset
new_df = pd.concat([legit_new_sample, fraud], axis=0)
new_df.shape
```

```python
new_df.head()
```

```python
new_df.tail()
```

```python
new_df.Class.value_counts()
```
